```python
import requests

def get_place_details(place_id):
    api_key = "AIzaSyDuREsWIzPRC5WdYSDQkxwVQDQAx8L3PeI"
    url = "https://maps.googleapis.com/maps/api/place/details/json"
    
    params = {
        "place_id": place_id,
        "fields": "name,rating,review",
        "key": api_key
    }
    
    response = requests.get(url, params=params)
    return response.json()

# Example
place_details = get_place_details("ChIJN1t_tDeuEmsRUsoyG83frY4")
for review in place_details.get("result", {}).get("reviews", []):
    print(f"Author: {review['author_name']}")
    print(f"Rating: {review['rating']}")
    print(f"Review: {review['text']}")
    print("-" * 50)
```
