```python
import pdfplumber
with pdfplumber.open("pdf.pdf") as pdf:
    for page_number, page in enumerate(pdf.pages):
        # Extract images from the page
        for image in page.images:
            print(f"Image found on page {page_number + 1}")
            
            print(image)
```
