## Basic Conversation with ConversationBufferMemory

```python
from langchain.chains.conversation.memory import ConversationBufferMemory
from langchain_groq import ChatGroq
from langchain.chains import ConversationChain

import os
from dotenv import load_dotenv
```

```python
load_dotenv()
os.environ["LANGCHAIN_API_KEY"] = os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

## Get Started

```python
from langchain.memory import ConversationBufferMemory
# You can also import it like this

memory = ConversationBufferMemory()
```

```python
memory.chat_memory.add_user_message("hi")
memory.chat_memory.add_ai_message("Hello! How are you?")
```

```python
memory.chat_memory.messages
```

```python
memory.load_memory_variables({})

# Bydefault memory return "history" key
```

```python
memory = ConversationBufferMemory(memory_key="chat_history")
# We can set name for memory key

memory.chat_memory.add_user_message("hi!")
memory.chat_memory.add_ai_message("Hello! How are you?")
```

```python
memory.load_memory_variables({})
```

```python
llm = ChatGroq(model="llama-3.3-70b-versatile")

memory = ConversationBufferMemory()

conversation = ConversationChain(llm = llm,
                                verbose=True,
                                memory=memory)
```

```python
print(conversation.memory.buffer)
```

```python
history = conversation.memory.buffer
print(len(history))
```

```python
conversation.predict(input="Hi there! Im sarawar")
```

```python
conversation.predict(input="Can you talk with me like my friend?")
```

```python
conversation.memory.buffer
```

```python
history = conversation.memory.buffer
print(len(history))
```

```python
for message in history:
    print(message, end="")
```

```python
if isinstance(history, str):
    history = history.split("\n")
```

```python
for message in history:
    message = message.strip()
    if message.startswith("Human:"):
        message = message.replace("Human: ", "", 1)
        print(message)
```

## Basic Conversation with ConversationBufferWindowMemory

```python
from langchain.memory import ConversationBufferWindowMemory
```

```python
# k=2 means memory will keep 2 recent interactions
window_memory = ConversationBufferWindowMemory(k=2)

conversation = ConversationChain(
    llm=llm,
    verbose=True,
    memory=window_memory
)
```

```python
conversation.predict(input="Hey! Im sarawar")
```

```python
conversation.predict(input="Im here to know about Conversation Buffer Window Memory in langchain")
```

```python
conversation.predict(input="So do you have any friends?")
```

# Conversation with Conversation Knowledge graph

```python
from langchain.chains.conversation.memory import ConversationKGMemory
from langchain.prompts.prompt import PromptTemplate
```

```python
template = """The following is a friendly conversation between a human and an AI. The AI is talkative and provides lots of specific details from its context. 
If the AI does not know the answer to a question, it truthfully says it does not know. The AI ONLY uses information contained in the "Relevant Information" section and does not hallucinate.

Relevant Information:

{history}

Conversation:
Human: {input}
AI:"""
prompt = PromptTemplate(
    input_variables=["history", "input"], template=template
)
```

```python
conversation_kg = ConversationChain(
    llm=llm,
    prompt=prompt,
    verbose=True,
    memory=ConversationKGMemory(llm=llm)
)
```

```python
conversation_kg.predict(input="Hi there! I am Sarawar")
```

```python
conversation_kg.predict(input="My TV is broken and I need some customer assistance")
```

```python
# Network entity graph
import networkx as nx
import matplotlib.pyplot as plt

# nx.draw(conversation_kg.memory.kg, with_labels=True)
# plt.show()

print(conversation_kg.memory.kg)
print(conversation_kg.memory.kg.get_triples())
```

## Entity Memory

```python
from langchain.memory import ConversationEntityMemory
from langchain.chains.conversation.prompt import ENTITY_MEMORY_CONVERSATION_TEMPLATE
from pydantic import BaseModel
from typing import List, Dict, Any
```

```python
conversation = ConversationChain(
    llm=llm, 
    verbose=True,
    prompt=ENTITY_MEMORY_CONVERSATION_TEMPLATE,
    memory=ConversationEntityMemory(llm=llm)
)
```
